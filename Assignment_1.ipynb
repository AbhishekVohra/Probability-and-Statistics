(2) Write a program in R to find factorial of a number by taking input from user. Please print error message if the input number is negative.

```python
f%%R
var = readline(prompt = "Enter a number")
FactVar = as.integer(var)

if (FactVar < 0){
  cat ("ERROR : input number is negative")
} else{
  i = 1
  product = 1
  
  while (i <= FactVar){
    product = product * i
    i = i+1
  }
  
  cat ("Factorial = ")
  cat (product)
}
```

(3) Write a program to write first n terms of a Fibonacci sequence. You may take n as an input from the user.


```python
%%R
var1 = readline(prompt = "Enter a number ")
newVar = as.integer(var1) # converts the input to integer which is originally a string

refVector = rep(1,newVar) # creates a vector which has newVar times value = '1'
# rep means repetition

for (x in 1:newVar){
  if (x == 1 || x == 2){ # first 2 elements of fibonacci series are 1
    cat(1)
    cat(" ")
  }else{
    refVector[x] = refVector[x-1] + refVector[x-2] # formula for fibonacci series
    cat(refVector[x], " ")
  }
}


```

(4) Write an R program to make a simple calculator which can add, subtract, multiply and divide.


```python
%%R
variable1 = readline(prompt = "Enter the first number : ")
newvar1 = as.integer(variable1)

variable2 = readline(prompt = "Enter the second number : ")
newvar2 = as.integer(variable2)

variable3 = readline(prompt = "Enter the operation \n + -> 'Addition' \n - -> 'Subtraction' \n * -> 'Multiplication' \n / -> 'Division' \n % -> 'Modulo' \n ^ -> 'Power' : ")

# using switch case statement for the calculator
# no default argument case is available in R in switch case

result = switch(
  variable3,
  "+"=cat("Addition = ", newvar1+newvar2),
  "-"=cat("Subtraction = ", newvar1-newvar2),
  "*"=cat("Multiplication = ", newvar1*newvar2),
  "/"=cat("Division = ", newvar1/newvar2),
  "%"=cat("Modulo = ", newvar1%%newvar2),
  "^"=cat("Power = ", newvar1^newvar2)
)

```

(5) Explore plot, pie, barplot etc. (the plotting options) which are built-in functions in R.

```python
%%R
x = c(1,2,3,4,5)
y = c(10,23,54,44,12)

# plot

# col => color
# cex => refers to the symbol size used for plotting
# pch => refers to the type of symbol used for plotting (here inverted triangle is used)
plot(x,y,col="red", cex=2, pch=25)
# type => refers to type of plot (here type="l" means type = line)
plot(x,y,type="l", col="blue")


# pie chart

labels=c("Audi", "Mercedes", "BMW", "VW", "Porsche")
# round to 1 decimal place
piepercentage = round(100 * x/sum(x), 1)
# main => main name of the chart
# rainbow((length(x))) => creates a rainbow color pallet which x different colors
pie(x,labels=piepercentage, main="Car Pie Chart", col=rainbow(length(x)))
# below line of code describes the position of the legend
legend("bottomleft", labels,cex = 1.0, fill = rainbow(length(x)))


# Bar plot

labels = c("Audi", "Mercedes", "BMW", "Porsche", "VW")
# names.arg => vector which appears under each bar
# xlabel/ylabel => names of the x and y axis respectively
barplot(x, names.arg=labels, xlab="Model", ylab="Number", col=rainbow(5), main="Car data")
# horiz=TRUE => makes a horizontal bar plot
barplot(x, names.arg=labels,horiz=TRUE, xlab="Model", ylab="Number", col=rainbow(5), main="Car data")
```
