Q1.<br> Consider that X is the time (in minutes) that a person has to wait in order to take a flight. <br>
If each flight takes off each hour X ~ U(0, 60). Find the probability that <br>
(a) waiting time is more than 45 minutes, and <br>
(b) waiting time lies between 20 and 30 minutes.

```r
# punif gives the distribution function
# lower.tail => logical; if TRUE, probabilties are P[X<=x]
x1_punif <- punif(45,0,60,lower.tail= FALSE)
x2_punif <- punif(30,0,60,lower.tail = TRUE) - punif(20,0,60,lower.tail = TRUE)
print(x1_punif)
print(x2_punif)
```

Q2. <br>
The time (in hours) required to repair a machine is an exponential distributed random <br>
variable with parameter λ = 1/2. <br>
(a) Find the value of density function at x = 3. <br>
(b) Plot the graph of exponential probability distribution for 0 ≤ x ≤ 5. <br>
(c) Find the probability that a repair time takes at most 3 hours. <br>
(d) Plot the graph of cumulative exponential probabilities for 0 ≤ x ≤ 5. <br>
(e) Simulate 1000 exponential distributed random numbers with λ = ½ and plot the 
simulated data.


```r
# dexp() function returns the corresponding values of the exponential density for an input vector of quantiles.
# dexp(x_dexp, rate)
x1<-dexp(3,1/2)
print(x1)

x<-c(0,1,2,3,4,5)
y<-dexp(x,1/2)
plot(y,type='o')

# pexp() function returns the corresponding values of the exponential cumulative distribution function for an input vector of quantiles.
# pexp(x_pexp, rate )
x2<-pexp(3,1/2,lower.tail = TRUE)
print(x2)

y1<- pexp(x,1/2)
plot(y1,type='o')

# The use of set. seed is to make sure that we get the same results for randomization.
set.seed(500) 

# rexp() function is used to simulate a set of random numbers drawn from the exponential distribution.
N <- 1000
y_rexp <- rexp(N, rate = 1/2)
plot(density(y_rexp))
```

Q3.<br>
The lifetime of certain equipment is described by a random variable X that follows <br>
Gamma distribution with parameters α = 2 and β = 1/3. <br>
(a) Find the probability that the lifetime of equipment is <br>&nbsp;&nbsp;(i) 3 units of time, and <br>&nbsp;&nbsp;(ii) 
at least 1 unit of time.<br>
(b) What is the value of c, if P(X ≤ c) ≥ 0.70? (Hint: try quantile function qgamma())

```r
# R function dgamma(x, reate) is the probability of interval x until the αth successful event when successful events occur with rate = 1/θ. 
x1<- dgamma(3,shape = 2,scale = 1/3)
print(x1)

# pgamma(q, shape, rate) – finds the value of the cumulative density function of a gamma distribution with certain shape and rate parameters. 
x2<- pgamma(1,shape = 2,scale = 1/3,lower.tail = FALSE)
print(x2)

# qgamma(p, shape, rate) – finds the value of the inverse cumulative density function of a gamma distribution with certain shape and rate parameters.
c_val <- qgamma(0.7,shape =2,scale = 1/3)
print(c_val)
```
