Q1. <br>
Use the rt(n, df) function in r to investigate the t-distribution for n = 100 and df = n − 1 and plot
the histogram for the same.

Q2. <br>
Use the rchisq(n, df) function in r to investigate the chi-square distribution with n = 100 and
df = 2, 10, 25

```r
# R function rchisq(n, df) returns n random numbers from the chi-square distribution. 
df <- c(2,10,25)
n <- 100
k <-length(df)
# writeLines() is used to write the sequence of multiple lines to the text file
for (i in 1:k) {
  print(rchisq(n,df[i]))
  writeLines("\n")
}
```

Q3. <br>
Generate a vector of 100 values between -6 and 6. Use the dt() function in r to find the values of a
t-distribution given a random variable x and degrees of freedom 1,4,10,30. Using these values plot
the density function for students t-distribution with degrees of freedom 30. Also shows a comparison
of probability density functions having different degrees of freedom (1,4,10,30).

```r
# Seq(): The seq function in R can generate the general or regular sequences from the given inputs. seq(from, to, by, length.out, along.with).
x <- seq(-6, 6, length = 100)

df = c(1,4,10,30)
colour = c("red", "orange", "green", "yellow","black")

# dnorm is the R function that calculates the p. d. f. f of the normal distribution. 
plot(x, dnorm(x), type = "l", lty = 2, xlab = "t-value", ylab = "Density",
     main = "Comparison of t-distributions", col = "black")

for (i in 1:4){
  lines(x, dt(x, df[i]), col = colour[i])
}

legend("topright", c("df = 1", "df = 4", "df = 10", "df = 30", "normal"),
       col = colour, title = "t-distributions", lty = c(1,1,1,1,2))
```

Q4. <br>
 Write a r-code <br>
(i) To find the 95th percentile of the F-distribution with (10, 20) degrees of freedom.<br>
(ii) To calculate the area under the curve for the interval [0, 1.5] and the interval [1.5, +∞) of
a F-curve with v1 = 10 and v2 = 20 (USE pf()). <br>
(iii) To calculate the quantile for a given area (= probability) under the curve for a F-curve 
with v1 = 10 and v2 = 20 that corresponds to q = 0.25, 0.5, 0.75 and 0.999. (use the qf()) <br>
(iv) To generate 1000 random values from the F-distribution with v1 = 10 and v2 = 20 (use
rf())and plot a histogram.

```r
# We use the qf() to calculate the quantile for a given area (= probability) under the curve for a F-curve
A <- qf(.95,df1=10,df2=20)
print(A)

# We use the pf() to calculate the area under the curve for the interval
B1 <- pf(1.5,df1=10,df2=20,lower.tail = TRUE)
B2 <- pf(1.5,df1=10,df2=20,lower.tail = FALSE)
print(B1)
print(B2)

# We use the qf() to calculate the quantile for a given area (= probability) under the curve for a F-curve
C1 = qf(0.25,df1=10,df2=20)
C2 = qf(0.5,df1=10,df2=20)
C3 = qf(0.75,df1=10,df2=20)
C4 = qf(0.999,df1=10,df2=20)
print(C1)
print(C2)
print(C3)
print(C4)

# We use rf()function to generate 100,000 random values from the F-distribution
D <- rf(n = 1000, df1=10,df2=20)
hist(D)
print(D)
```
